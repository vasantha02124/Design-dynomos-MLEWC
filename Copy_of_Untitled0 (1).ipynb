```python
#handling missing data or cleaning data
#step3 handing missing data and NaN values
#scenario1 : missing values in numerical column
#scenario 2: fill the empty value in highly repeated values
#scenario3: removing data set from 2 rows

import numpy as np
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])
```

```python
X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
np.remainder='passthrough'
X=ct.fit_transform(X)
```

```python
#onehotencoder is only usefull for input column
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
# step 5 :slipt the data in to training and testing data.
from sklearn. model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```
